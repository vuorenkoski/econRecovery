# View WB, IMF and Covid data combined

Data is previoiusly saved locally in CSV format, where columns are 'Country or Area', 'Item', 'Year', 'Value'.

```python
import pandas as pd

df=pd.read_csv('IMF_statistics.csv')
df=df.append(pd.read_csv('WB_statistics.csv'))
df=df.append(pd.read_csv('Covid_statistics.csv'))
print('Number of datapoints:',df.shape[0])
```

```python
print('Indicators in data')
indicators=df.Item.unique()
for x in indicators:
    print(' ',x)
    print('    years:',df[df.Item==x].Year.unique())
```

```python
df.to_csv('Combined_statistics.csv', index=False)

countries=df['Country or Area'].unique()
print('Number of countries or areas in data:',len(countries))
df[df['Country or Area']=='Finland']
```

```python
ccodes=pd.read_csv('countries.csv')[['Code2','Code3','Country or Area']]
ccodes.head()
```

```python
df=df.drop(columns={'Country or Area'})
```

```python
df2=pd.merge(df, ccodes, on=['Code3'], how='left')
df2.head(50)
df2.shape
```

```python
df2.to_csv('Combined_statistics.csv', index=False)
```
